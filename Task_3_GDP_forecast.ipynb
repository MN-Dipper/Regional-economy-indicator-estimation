## 找超参数

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
import xgboost as xgb
from itertools import product
from tqdm import tqdm 
import warnings
from sklearn.exceptions import ConvergenceWarning
from tqdm import tqdm


def mean_absolute_percentage_error(y_true, y_pred): 
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

def EN_CV(X_train, y_train, param_grid, stratifier):
    best_r2_test = -np.inf
    best_params = None
    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    pbar = tqdm(total=skf.get_n_splits() * len(param_grid['alpha']) * len(param_grid['l1_ratio']))

    for train_index, test_index in skf.split(X_train, stratifier):
        X_train_fold, X_test_fold = X_train[train_index], X_train[test_index]
        y_train_fold, y_test_fold = y_train[train_index], y_train[test_index]

        scaler_X = StandardScaler()
        X_train_scaled = scaler_X.fit_transform(X_train_fold)
        X_test_scaled = scaler_X.transform(X_test_fold)
        
        scaler_y = StandardScaler()
        y_train_scaled = scaler_y.fit_transform(y_train_fold.reshape(-1, 1)).ravel()
        y_test_scaled = scaler_y.transform(y_test_fold.reshape(-1, 1)).ravel()
        
        for alpha in param_grid['alpha']:
            for l1_ratio in param_grid['l1_ratio']:
                model = ElasticNet(alpha=alpha, l1_ratio=l1_ratio, max_iter=10000)
                model.fit(X_train_scaled, y_train_scaled)
                y_test_pred_scaled = model.predict(X_test_scaled)
                y_test_pred = scaler_y.inverse_transform(y_test_pred_scaled.reshape(-1, 1)).ravel()
                r2_test = r2_score(y_test_fold, y_test_pred)

                if r2_test > best_r2_test:
                    best_r2_test = r2_test
                    best_params = {'alpha': alpha, 'l1_ratio': l1_ratio}

                pbar.update(1)

    pbar.close()
    return best_r2_test, best_params
```

```python
def merge_and_sort_dataframes(df1, df2, ascending=False):
    df1.rename(columns={'Feature': 'item_first_cate_name'}, inplace=True)

    merged_df = pd.merge(df1[['item_first_cate_name', 'Weight']],
                         df2[['item_first_cate_name', 'nums']],
                         on='item_first_cate_name', how='outer')
    
    merged_df['Weight_abs'] = merged_df['Weight'].abs() 
    sorted_merged_df = merged_df.sort_values(by='Weight_abs', ascending=ascending)
    
    sorted_merged_df['Weight_Rank'] = sorted_merged_df['Weight'].abs().rank(ascending=ascending, method='average')  
    sorted_merged_df['Nums_Rank'] = sorted_merged_df['nums'].rank(ascending=False, method='average')  

    return sorted_merged_df
```

# ElasticNet使用

```python
from sklearn.model_selection import KFold
import numpy as np
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import xgboost as xgb
import numpy as np
import pandas as pd
from sklearn.model_selection import StratifiedShuffleSplit, StratifiedKFold
from sklearn.preprocessing import StandardScaler


def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    

def EN_evaluate(data, model_name, param_grid, year):
    random_seed = 51
    data_1 = data[data['year'] == year]
    data_1 = data_1.reset_index(drop=True)
    metrics = []
    best_r2_test = -np.inf  
    best_weights = None  
    feature_names = data.columns[3:-1]
    X = data_1.iloc[:, 3:-1].values
    y = data_1.iloc[:, -1].values
    
    data_2 = data[data['year'] == year + 1]
    data_2 = data_2.reset_index(drop=True)
    X_1 = data_2.iloc[:, 3:-1].values
    y_1 = data_2.iloc[:, -1].values
    
    stratifier = data_1['city_level']
    
    X_train, X_test, y_train, _, strat_train, strat_test = train_test_split(X, y, stratifier, test_size=0.3, stratify=stratifier, random_state=random_seed)
    _, _, _, y_test, _, _ = train_test_split(X_1, y_1, stratifier, test_size=0.3, stratify=stratifier, random_state=random_seed)

    
    r2_test, best_params = EN_CV(X_train, y_train, param_grid, strat_train)

    scaler_X = StandardScaler()
    X_train_scaled = scaler_X.fit_transform(X_train)
    X_test_scaled = scaler_X.transform(X_test)

    scaler_y = StandardScaler()
    y_train_scaled = scaler_y.fit_transform(y_train.reshape(-1, 1)).ravel()
    y_test_scaled = scaler_y.transform(y_test.reshape(-1, 1)).ravel()
                    
    xg_reg = ElasticNet(alpha=best_params['alpha'], l1_ratio=best_params['l1_ratio'], max_iter=10000)
    
    xg_reg.fit(X_train_scaled, y_train_scaled)

    y_train_pred_scaled = xg_reg.predict(X_train_scaled)
    y_test_pred_scaled = xg_reg.predict(X_test_scaled)
    

    y_train_pred = scaler_y.inverse_transform(y_train_pred_scaled.reshape(-1, 1)).ravel()
    y_test_pred = scaler_y.inverse_transform(y_test_pred_scaled.reshape(-1, 1)).ravel()

    r2_test = r2_score(y_test, y_test_pred)
    
    best_r2_test = r2_test
    best_weights = xg_reg.coef_

    fold_metrics = {
        'RMSE_train': np.sqrt(mean_squared_error(y_train, y_train_pred)),
        'MAPE_train': mean_absolute_percentage_error(y_train, y_train_pred),
        'MAE_train': mean_absolute_error(y_train, y_train_pred),
        'R2_train': r2_score(y_train, y_train_pred),
        
        'RMSE_test': np.sqrt(mean_squared_error(y_test, y_test_pred)),
        'MAPE_test': mean_absolute_percentage_error(y_test, y_test_pred),
        'MAE_test': mean_absolute_error(y_test, y_test_pred),
        'R2_test': r2_score(y_test, y_test_pred)
    }
        
        
    metrics.append({**fold_metrics})

    results_df = pd.DataFrame(metrics)

    results_df.insert(0, 'Data', model_name)

    best_weights_df = pd.DataFrame({'Feature': feature_names, 'Weight': best_weights})
    return best_weights_df, results_df,best_params
```

```python
def merge_and_sort_dataframes(df1, df2, ascending=False):
    df1.rename(columns={'Feature': 'item_first_cate_name'}, inplace=True)

    merged_df = pd.merge(df1[['item_first_cate_name', 'Weight']],
                         df2[['item_first_cate_name', 'nums']],
                         on='item_first_cate_name', how='outer')
    

    merged_df['Weight_abs'] = merged_df['Weight'].abs()  
    sorted_merged_df = merged_df.sort_values(by='Weight_abs', ascending=ascending)
    
    sorted_merged_df['Weight_Rank'] = sorted_merged_df['Weight'].abs().rank(ascending=ascending, method='average')  
    sorted_merged_df['Nums_Rank'] = sorted_merged_df['nums'].rank(ascending=False, method='average')  

    return sorted_merged_df
```

```python
def process_weights_data(EN_weights, year, experiment_name, grouped_df):

    filtered_weights = EN_weights[EN_weights['Experiment'].str.contains(year)]
    filtered_weights = filtered_weights[filtered_weights['Experiment'].str.contains(experiment_name)]
    
    merged_data = merge_and_sort_dataframes(filtered_weights, grouped_df)
    
    result_df = merged_data[['item_first_cate_name', 'Weight', 'Weight_Rank', 'nums', 'Nums_Rank']]

    result_df.loc[:, 'nums'] = result_df['nums'].round(1)

    return result_df
```

```python
from sklearn.linear_model import ElasticNet
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

import numpy as np
import pandas as pd

def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

def EN_evaluate_incremental(data, model_name, param_grid, year, importance_order):
    data = data[data['year'] == year]
    data = data.reset_index(drop=True)
    
    # 获取特征名称对应的索引
    feature_names = data.columns[3:-1]  # 假设特征开始于第四列并结束于倒数第二列
    # print("Feature columns in data:", feature_names)  # 打印实际的特征列名称
    
    # 验证并转换特征名称为索引
    importance_indices = []
    for name in importance_order:
        if name in feature_names:
            importance_indices.append(feature_names.get_loc(name))
        else:
            print(f"Feature name '{name}' not found in data columns.")

    # print("Converted importance indices:", importance_indices)  # 打印转换后的索引

    metrics = []
    X = data.iloc[:, 3:-1].values
    y = data.iloc[:, -1].values
    stratifier = data['city_level']

    # print("Shape of X:", X.shape)  # 打印 X 的形状

    try:
        X = X[:, importance_indices]
    except IndexError as e:
        print("Index error:", e)
        return None

    # 分割数据
    X_train, X_test, y_train, y_test, strat_train, strat_test = train_test_split(X, y, stratifier, test_size=0.3, stratify=stratifier, random_state=21)

    scaler_X = StandardScaler()
    scaler_y = StandardScaler()
    
    # 逐步增加特征
    for i in range(1, len(importance_indices) + 1):
        X_train_subset = scaler_X.fit_transform(X_train[:, :i])
        X_test_subset = scaler_X.transform(X_test[:, :i])

        r2_test, best_params = EN_CV(X_train[:, :i], y_train, param_grid, strat_train)

        # print("Shape of X_train_subset:", X_train_subset.shape)  # 打印 X 的形状
        y_train_scaled = scaler_y.fit_transform(y_train.reshape(-1, 1)).ravel()
        y_test_scaled = scaler_y.transform(y_test.reshape(-1, 1)).ravel()

        # 使用ElasticNet训练模型
        enet = ElasticNet(alpha=best_params['alpha'], l1_ratio=best_params['l1_ratio'], max_iter=10000)
        enet.fit(X_train_subset, y_train_scaled)

        # 预测
        y_train_pred = scaler_y.inverse_transform(enet.predict(X_train_subset).reshape(-1, 1)).ravel()
        y_test_pred = scaler_y.inverse_transform(enet.predict(X_test_subset).reshape(-1, 1)).ravel()

        # 计算性能指标
        r2_train = r2_score(y_train, y_train_pred)
        r2_test = r2_score(y_test, y_test_pred)

        metrics.append({
            'Num_features': i,
            'R2_train': r2_train,
            'R2_test': r2_test
        })

    # 将性能指标的列表转换成DataFrame
    results_df = pd.DataFrame(metrics)
    results_df.insert(0, 'Data', model_name)
    return results_df
```

```python
import warnings
warnings.filterwarnings("ignore", category=UserWarning)

def XGB_CV(X_train, y_train, param_grid, stratifier):
    best_r2_test = -np.inf
    best_params = None
    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    
    # 更新总进度条计数
    pbar = tqdm(total=skf.get_n_splits() * len(param_grid['learning_rate']) * len(param_grid['max_depth']) * len(param_grid['n_estimators']) * len(param_grid['alpha']))
    # * len(param_grid['colsample_bytree'])
    for train_index, test_index in skf.split(X_train, stratifier):
        X_train_fold, X_test_fold = X_train[train_index], X_train[test_index]
        y_train_fold, y_test_fold = y_train[train_index], y_train[test_index]

        scaler_X = StandardScaler()
        X_train_scaled = scaler_X.fit_transform(X_train_fold)
        X_test_scaled = scaler_X.transform(X_test_fold)

        scaler_y = StandardScaler()
        y_train_scaled = scaler_y.fit_transform(y_train_fold.reshape(-1, 1)).ravel()
        y_test_scaled = scaler_y.transform(y_test_fold.reshape(-1, 1)).ravel()

        for lr in param_grid['learning_rate']:
            for depth in param_grid['max_depth']:
                for n_est in param_grid['n_estimators']:
                    for alpha in param_grid['alpha']:
                        # for colsample in param_grid['colsample_bytree']:
                        model = xgb.XGBRegressor(
                            learning_rate=lr, 
                            max_depth=depth, 
                            n_estimators=n_est, 
                            reg_alpha=alpha, 
                            # colsample_bytree=colsample
                        )
                        model.fit(X_train_scaled, y_train_scaled)
                        y_test_pred_scaled = model.predict(X_test_scaled)
                        y_test_pred = scaler_y.inverse_transform(y_test_pred_scaled.reshape(-1, 1)).ravel()
                        r2_test = r2_score(y_test_fold, y_test_pred)

                        if r2_test > best_r2_test:
                            best_r2_test = r2_test
                            best_params = {
                                'learning_rate': lr,
                                'max_depth': depth,
                                'n_estimators': n_est,
                                'alpha': alpha,
                                # 'colsample_bytree': colsample
                            }
                        pbar.update(1)

    pbar.close()
    return best_r2_test, best_params
```

```python
from sklearn.model_selection import KFold
import numpy as np
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import xgboost as xgb
import numpy as np
import pandas as pd
from sklearn.model_selection import StratifiedShuffleSplit, StratifiedKFold
from sklearn.preprocessing import StandardScaler


def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    
def XGB_evaluate(data, model_name, param_grid, year):

    data = data[data['year'] == year]
    data = data.reset_index(drop=True)
    metrics = []
    best_r2_test = -np.inf  # 初始化最高R2_test为无限小
    best_weights = None  # 用于存储最高R2_test对应的权重
    feature_names = data.columns[3:-1]
    X = data.iloc[:, 3:-1].values
    y = data.iloc[:, -1].values

    stratifier = data['city_level']
    
    X_train, X_test, y_train, y_test, strat_train, strat_test = train_test_split(X, y, stratifier, test_size=0.3, stratify=stratifier, random_state=21)

    r2_test, best_params = XGB_CV(X_train, y_train, param_grid, strat_train)

    scaler_X = StandardScaler()
    X_train_scaled = scaler_X.fit_transform(X_train)
    X_test_scaled = scaler_X.transform(X_test)

    scaler_y = StandardScaler()
    y_train_scaled = scaler_y.fit_transform(y_train.reshape(-1, 1)).ravel()
    y_test_scaled = scaler_y.transform(y_test.reshape(-1, 1)).ravel()
                    
    # 初始化和训练模型
    xg_reg = xgb.XGBRegressor(
        objective='reg:squarederror',
        # colsample_bytree=best_params['colsample_bytree'],
        learning_rate=best_params['learning_rate'],
        max_depth=best_params['max_depth'],
        alpha=best_params['alpha'],
        n_estimators=best_params['n_estimators']
    )
    xg_reg.fit(X_train_scaled, y_train_scaled)
    # 进行预测
    y_train_pred_scaled = xg_reg.predict(X_train_scaled)
    y_test_pred_scaled = xg_reg.predict(X_test_scaled)
    
    # 反归一化预测结果
    y_train_pred = scaler_y.inverse_transform(y_train_pred_scaled.reshape(-1, 1)).ravel()
    y_test_pred = scaler_y.inverse_transform(y_test_pred_scaled.reshape(-1, 1)).ravel()

    r2_test = r2_score(y_test, y_test_pred)
    
    if r2_test > best_r2_test:
        best_r2_test = r2_test
        best_weights = xg_reg.feature_importances_
    # 计算评估指标
    fold_metrics = {
        'RMSE_train': np.sqrt(mean_squared_error(y_train, y_train_pred)),
        'MAPE_train': mean_absolute_percentage_error(y_train, y_train_pred),
        'MAE_train': mean_absolute_error(y_train, y_train_pred),
        'R2_train': r2_score(y_train, y_train_pred),
        
        'RMSE_test': np.sqrt(mean_squared_error(y_test, y_test_pred)),
        'MAPE_test': mean_absolute_percentage_error(y_test, y_test_pred),
        'MAE_test': mean_absolute_error(y_test, y_test_pred),
        'R2_test': r2_score(y_test, y_test_pred)
    }
        
        
    metrics.append({**fold_metrics})
    # 将性能指标的列表转换成DataFrame
    results_df = pd.DataFrame(metrics)
    # 计算平均性能指标
    results_df.insert(0, 'Data', model_name)

    # 在process_and_evaluate函数内
    best_weights_df = pd.DataFrame({'Feature': feature_names, 'Weight': best_weights})
    return best_weights_df, results_df, best_params
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import xgboost as xgb

def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    

def process_and_evaluate(data, model_name):
    metrics = []

    # 数据准备
    X = data.iloc[:, 2:-1].values
    y = data.iloc[:, -1].values
    groups = data.iloc[:, 0].values  # 使用城市ID作为分组依据
    feature_names = data.columns[2:-1]

    gkf = GroupKFold(n_splits=5)

    for train_index, test_index in gkf.split(X, y, groups):
        # 分割数据
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]     
        
        # 数据标准化
        scaler_X = StandardScaler()
        X_train_scaled = scaler_X.fit_transform(X_train)
        X_test_scaled = scaler_X.transform(X_test)

        scaler_y = StandardScaler()
        y_train_scaled = scaler_y.fit_transform(y_train.reshape(-1, 1)).ravel()
        y_test_scaled = scaler_y.transform(y_test.reshape(-1, 1)).ravel()

        # 初始化和训练模型
        xg_reg = xgb.XGBRegressor(objective='reg:squarederror', colsample_bytree=0.8, learning_rate=0.01, max_depth=500, alpha=0.3, n_estimators=1000)
        xg_reg.fit(X_train_scaled, y_train_scaled)

        # 进行预测
        y_train_pred_scaled = xg_reg.predict(X_train_scaled)
        y_test_pred_scaled = xg_reg.predict(X_test_scaled)

        # 反归一化预测结果
        y_train_pred = scaler_y.inverse_transform(y_train_pred_scaled.reshape(-1, 1)).ravel()
        y_test_pred = scaler_y.inverse_transform(y_test_pred_scaled.reshape(-1, 1)).ravel()

        # 计算评估指标
        fold_metrics = {
            'RMSE_train': np.sqrt(mean_squared_error(y_train, y_train_pred)),
            'MAPE_train': mean_absolute_percentage_error(y_train, y_train_pred),
            'MAE_train': mean_absolute_error(y_train, y_train_pred),
            'R2_train': r2_score(y_train, y_train_pred),
            'RMSE_test': np.sqrt(mean_squared_error(y_test, y_test_pred)),
            'MAPE_test': mean_absolute_percentage_error(y_test, y_test_pred),
            'MAE_test': mean_absolute_error(y_test, y_test_pred),
            'R2_test': r2_score(y_test, y_test_pred)
        }
        
        metrics.append({**fold_metrics})
    # 将性能指标的列表转换成DataFrame
    results_df = pd.DataFrame(metrics)
    # 计算平均性能指标
    results_df.insert(0, 'Data', model_name)

    return results_df
```

```python
dataframes = {'EN 品类+人口': data_exp_1, 'EN 只用人口': data_exp_2,'EN 只用品类': data_exp_3}
all_results = pd.DataFrame()
all_weight = pd.DataFrame()
for name, df in dataframes.items():
    result_df = process_and_evaluate(df, name)
    all_results = pd.concat([all_results, result_df], ignore_index=True)

all_results
```
